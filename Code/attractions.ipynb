```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter


file_path = "attractions.csv"
df = pd.read_csv(file_path, skiprows=2) # Skip header
df.set_index("Unnamed: 0", inplace=True)
df.index.name = "Attraction"

# clean data
df = df.applymap(lambda x: int(str(x).replace(",", "")))
df = df / 1000

# creating heatmap
plt.figure(figsize=(12, 6))
ax = sns.heatmap(df, annot=True, fmt=".3f", cmap="YlGnBu", linewidths=0.5,
                 cbar_kws={'label': 'Visitors (in millions)'})
cbar = ax.collections[0].colorbar
cbar.ax.yaxis.set_major_formatter(ScalarFormatter(useOffset=False))
cbar.ax.tick_params(labelsize=10)
plt.title("Number of Visitors for London Attractions \n (2019-2023)", fontsize=16, fontweight='bold')
plt.xlabel("Year", fontsize=12, fontweight='bold')
plt.ylabel("Attractions", fontsize=12, fontweight='bold')


plt.tight_layout()
```
