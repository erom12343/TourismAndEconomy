```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean the data
df = pd.read_csv("total_tourists.csv")
clean_df = df.iloc[2:].copy()
clean_df.columns = ['Year', 'Tourist_Visits']
clean_df['Year'] = clean_df['Year'].astype(int)
clean_df['Tourist_Visits'] = clean_df['Tourist_Visits'].astype(float)

# plotting data
sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=clean_df, x='Year', y='Tourist_Visits', marker='o')
plt.fill_between(clean_df['Year'].values, clean_df['Tourist_Visits'].values, color='skyblue', alpha=0.4)
plt.title('International Tourist Visits to London (2011â€“2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Visits (millions)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
